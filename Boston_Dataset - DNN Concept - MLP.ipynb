```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
# iris
# titanic
```

```python
X, y = load_boston(return_X_y = True)
```

```python
print(X.shape)
print(y.shape)
```

```python
pd.DataFrame(X)
```

```python
# splitting the data into training and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Deep Learning - Nearual Network

# MultiLayer Perceptron

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# Sequential API - Fully connected layer (one by one)
dnn = tf.keras.models.Sequential()
```

```python
# adding first hideen layer into the input layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# adding 2nd hideen layer into the input layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Adding output layer 
dnn.add(tf.keras.layers.Dense(units=1))
```

```python
# Train the Deep Learning Model for regression problem

dnn.compile(optimizer ='adam' , loss='mean_squared_error' )
```

```python
# predict/evaluate the model

dnn.fit(X_train, y_train, batch_size= 32, epochs= 500)
```

```python
dnn.fit(X_train, y_train, batch_size= 32, epochs= 1000)
```

```python
# predict the price

y_pred = dnn.predict(X_test)
print(y_pred)
```
