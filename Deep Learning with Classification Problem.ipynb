```python
import numpy as np
import pandas as pd
import tensorflow as tf
# matplotlib - ML and Tensorboard - DL
# numpy - ML and Tensorflow - DL
```

```python
# Importing the dataset 
dataset = pd.read_csv("C:\\Users\\Lenovo\\Desktop\\Churn_Modelling.csv")
dataset.head()
```

```python
dataset['Geography'] = dataset['Geography'].astype('category')
dataset['Geography'] = dataset['Geography'].cat.codes
```

```python
dataset['Gender'] = dataset['Gender'].astype('category')
dataset['Gender'] = dataset['Gender'].cat.codes
dataset.head()
```

```python
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
pd.DataFrame(X).head()
```

```python
pd.DataFrame(y).head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1)
```

```python
# Building Deep Neural Network
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
dnn = tf.keras.models.Sequential()
```

```python
# Add first hidden layer

dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Add 2nd hidden layer

dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Adding output layer 
dnn.add(tf.keras.layers.Dense(units=1, activation = 'sigmoid'))
```

```python
# Train the Deep Learning Model for Classification problem

dnn.compile(optimizer ='adam' , loss='binary_crossentropy',metrics=['accuracy'] )
```

```python
# predict/evaluate the model

dnn.fit(X_train, y_train, batch_size= 32, epochs= 500)
```

```python
# predict/evaluate the model

dnn.fit(X_train, y_train, batch_size= 32, epochs= 100)
```

```python
# predict the test set result

y_pred = dnn.predict(X_test)
y_pred = (y_pred>0.5)
#y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
